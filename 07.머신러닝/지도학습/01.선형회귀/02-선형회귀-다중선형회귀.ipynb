```python
# 선형 회귀의 이해
# 원인과 결과의 관계를 선으로 그리는 작업
# 원인이 되는 데이터 (X) vs 결과를 나타내는 데이터(Y)
# 마법의 직선을 찾는 과정임(가장 훌륭한 예측선 찾기)

# 선형 회귀는 주어진 데이터로부터, 관계를 가장 잘 나타내는 직선을 그리는 일
```

```python
#선형회귀 모델 만들기
from sklearn.linear_model import LinearRegression

# 1. 회귀란 무엇인가?
import pandas as pd

df = pd.DataFrame({
    "study": [1,2,3,4,5,6,7,8],
    "sleep": [8,7,7,6,6,5,5,4],
    "homework": [1,1,2,2,3,3,4,4],
    "score": [50,55,60,65,72,78,85,90]
})

df

x = df[["study","sleep","homework"]]
y = df["score"] # 정답 = 실제 값

model = LinearRegression()
# x → (행, 열)
# y → (행,)
model.fit(x,y)

# score = 3.91 x study -1.3 x sleep + 2.58 x homework + 47.1
print(model.coef_) # [ 3.91666667 -1.33333333  2.58333333]
print(model.intercept_)

# 공부 6시간
# 수면 6시간
# 과제 3개

pred = model.predict([[6,6,3]])

print(pred)
```
