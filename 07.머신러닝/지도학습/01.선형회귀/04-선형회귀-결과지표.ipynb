```python
import pandas as pd

df = pd.DataFrame({
    "study":[1,2,3,4,5,6,7,8],
    "sleep":[8,7,7,6,6,5,5,4],
    "homework":[2,4,6,8,10,12,14,16],
    "score":[52,58,65,70,76,82,90,95]
})

print(df)
```

```python
# 2. 회귀모델 만들기

import statsmodels.api as sm

x = df[["study","sleep","homework"]]
y = df["score"]

x = sm.add_constant(x) # 1단계. 절편 추가
model = sm.OLS(y,x)    # 오차 제곱합을 최소로 하는 직선을 찾겠다는 뜻

result = model.fit()

print(result.summary()) 
# F-statistic:                     2103. - f분포
# strong multicollinearity problems or that the design matrix is singular.(Cond. No. - 3.05e+16) - 다중공선성 문제
# → 입력 변수(x)들끼리 너무 비슷해서 모델이 어떤 변수가 진짜 영향을 주는 지 구분하기 어려운 형태.
```
