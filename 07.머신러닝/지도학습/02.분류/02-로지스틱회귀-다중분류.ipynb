```python
import pandas as pd

df = pd.DataFrame({
    'study_time': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'grade':      [0, 0, 0, 1, 1, 1, 2, 2, 2]
})

df
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

x = df[["study_time"]]
y = df["grade"]

model.fit(x,y)

pred = model.predict(x)
print(pred)
# [0 0 0 1 1 1 2 2 2]

# 공부시간이 10시간이면 몇 등급인가?

pred = model.predict([[10],[4.5]]) # 10 시간일 때
print(pred)

print("클래스",model.classes_)
print("확률",model.predict_proba([[10]]))
```
