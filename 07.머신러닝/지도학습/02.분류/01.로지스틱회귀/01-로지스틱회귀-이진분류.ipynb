```python
import pandas as pd

df = pd.DataFrame({
    'study_time': [1, 2, 3, 4, 5, 6],
    'pass':       [0, 0, 0, 1, 1, 1]
})

df
```

```python
from sklearn.linear_model import LogisticRegression

# x, y 분리
x = df[["study_time"]]
y = df["pass"]

print(x.shape)
print(y.shape)

# 모델 생성
model = LogisticRegression()

# 학습
model.fit(x,y)

print(model.coef_)
print(model.intercept_)

# 예측
pred = model.predict(x)
print(pred)

pred = model.predict([[10]]) # 10 시간일 때
print(pred)

# 10시간일 때 12 일때
pred = model.predict([[10],[12]]) # 10 시간일 때
print(pred)
```

```python
# 운동시간으로 비만 여부 예측

# 운동시간(exercise)에 따라 비만 여부(obesity)를 예측한다.

# 0 → 비만
# 1 → 정상
import pandas as pd

data = pd.DataFrame({
    "exercise": [1,2,3,4,5,6,7,8],
    "obesity":  [0,0,0,0,1,1,1,1]
})

data

# 운동 10시간일때 비만 인지를 예측하시오.

from sklearn.linear_model import LogisticRegression

# x, y 분리
x = data[["exercise"]]
y = data["obesity"]

print(x.shape)
print(y.shape)

# 모델 생성
model = LogisticRegression()

# 학습
model.fit(x,y)

print(model.coef_)
print(model.intercept_)

# 예측
pred = model.predict([[10]])
print(pred)
```

```python
# 수면시간(sleep)에 따라 다음날 집중 가능 여부(focus)를 예측한다.

# 0 → 집중 못함
# 1 → 집중 잘함
import pandas as pd

data = pd.DataFrame({
    "sleep": [3,4,5,6,7,8,9,10],
    "focus": [0,0,0,0,1,1,1,1]
})

data

# 문제

# 아래 조건으로 로지스틱 회귀(LogisticRegression) 를 적용하시오.

# sleep을 X로 지정
# focus를 y로 지정
# 모델 학습 수행
# 수면시간이 11시간일 때 집중 가능 여부 예측
# predict_proba()를 사용하여 확률 출력

from sklearn.linear_model import LogisticRegression

# x, y 분리
x = data[["sleep"]]
y = data["focus"]

print(x.shape) # (8, 1)
print(y.shape) # (8,)

# 모델 생성
model = LogisticRegression()

# 학습
model.fit(x,y)

print(model.coef_) # [[1.16976606]]
print(model.intercept_) # [-7.60347386]

# 예측
pred = model.predict([[11]])
print(pred) # [1]

pred = model.predict_proba([[11]])
print(pred) # [[0.00514817 0.99485183]]
```
