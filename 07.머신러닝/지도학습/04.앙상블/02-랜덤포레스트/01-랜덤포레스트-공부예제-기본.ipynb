```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

df = pd.DataFrame({
    "study":[1,2,3,4,5,6,7,8],
    "sleep":[8,7,7,6,6,5,5,4],
    "homework":[1,2,3,4,5,6,7,8],
    "pass":[0,0,0,0,1,1,1,1]
})

X = df[["study","sleep","homework"]]
y = df["pass"]

# 배깅 적용
model = RandomForestClassifier(
    n_estimators=100,  # 몇 개 모델을 만들 것인가
    random_state=42    # 변하지 않게
)

# 학습
model.fit(X,y)

# 예측
pred = model.predict([[9,5,5]])

print(pred)
```
