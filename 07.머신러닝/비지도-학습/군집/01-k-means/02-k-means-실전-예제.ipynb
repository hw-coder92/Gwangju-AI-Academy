```python
# 구매횟수 / 구매금액

import pandas as pd
from sklearn.cluster import KMeans

df = pd.DataFrame({
    "횟수":[30,35,25,3,5,2],
    "금액":[100,120,90,10,15,8]
})

print(df)

model = KMeans(n_clusters = 2)

model.fit(df)

df["cluster"] = model.labels_

print(df)
print(model.cluster_centers_)

# 분류 결과
# 0 = VIP
# 1 = 일반고객
```

```python
import pandas as pd

df = pd.DataFrame({

    "시간":[
        150,140,160,
        80,90,100
    ],

    "평점":[
        9,10,8,
        3,4,2
    ]

})

print(df)

model = KMeans(n_clusters = 2)

model.fit(df)

df["cluster"] = model.labels_

print(df)
print(model.cluster_centers_)
```

```python
import pandas as pd
# 영화시청 시간과 평점
df = pd.DataFrame({
    "시간": [150,140,160,80,90,100],
    "평점": [9,10,8,3,4,2]
})

model = KMeans(n_clusters = 2)

model.fit(df)

df["cluster"] = model.labels_

print(model.cluster_centers_)
print(df)

# 응집도
print("응집도",model.inertia_)

# 실루엣 개수
from sklearn.metrics import silhouette_score,silhouette_samples

sil_score = silhouette_score(df[["시간","평점"]],model.labels_)

print("실루엣 스코어",sil_score)

# 시청시간이 길면 => 평점높음 => 긴 영화 선호
# 시청시간이 짧으면 => 평점 낮음 => 짧은(단편) 영화 선호
```
