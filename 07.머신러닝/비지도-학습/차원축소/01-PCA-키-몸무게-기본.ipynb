```python
# PCA (Principal Component Analysis) => 주성분 분석
# Principal → 주요한, 핵심적인
# Component → 구성 요소(축)
# Analysis → 분석

import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = np.array([
    [170, 65],
    [175, 70],
    [180, 75],
    [185, 80]
])

pca = PCA(n_components=1)

# 1. fit() => 데이터의 규칙만 학습,
x_pca = pca.fit_transform(X)

print("원본 데이터:\n",X)
print("PCA 결과:\n",x_pca)

plt.figure()
plt.scatter(X[:, 0], X[:, 1])
plt.title("Original Data")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
