```python
import matplotlib.pyplot as plt

x = [ i for i in range(1,6)]
y = [ i for i in range(1,6)]

plt.plot(x,y)
plt.title("첫번째 그래프")
plt.xlabel("x축")
plt.ylabel("y축")

#plt.show()

# 객체기반으로 그리기
# fig는 전체 도화지를 의미함
# ax는 그래프를 그릴 수 있는 곳

x = [ 1,2,3,4,5]
y = [ 1,2,3,4,5]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel("x축")
ax.set_ylabel("y축")

fig.set_dpi(300)
fig.savefig("first_plot.png")
```

```python
# 1. Figure 와 Axes 2개 생성 (1행 2열 구조)

fig , ax = plt.subplots(1,2,figsize=(10,4)) # 가로 10인치 세로 4인치
ax[0].plot([1,2,3,4],[1,2,3,4],color="blue",marker="o")
ax[0].set_title("first")
ax[0].set_xlabel("X axis",loc="right",labelpad=15, fontdict={'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14})
ax[0].set_ylabel("Y axis",loc="top",labelpad=20, fontdict={'family': 'serif', 'color': 'r', 'weight': 'normal', 'size': "xx-large"})

ax[1].plot([1,2,3,4],[1,2,3,4],color="green",marker="+")
ax[1].set_title("first")
ax[1].set_xlabel("X axis",loc="right",labelpad=15, fontdict={'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14})
ax[1].set_ylabel("Y axis",loc="top",labelpad=20, fontdict={'family': 'serif', 'color': 'r', 'weight': 'normal', 'size': "xx-large"})
```

```python
# 눈금선 표시 컨트롤
import numpy as np

x = np.arange(1,11)
y1 = x ** 2
y2 = x ** 3
y3 = x * 4

plt.plot(x, y1,'ro--', x, y2, 'gs--', x, y3, 'b^--')
plt.legend(['red', 'green', 'blue'], loc='best')

# 눈금선의 스타일 설정(옵션 설정)
# describes a sequence of 5 point and 1 point dashes separated by 2
# "길이가 5인 선, 그 다음 간격 2, 길이가 1인 선, 그 다음 간격 2"의 패턴이 반복되도록 설정
# which='both' = 주눈금선(major ticks)과 보조눈금선(minor ticks) 모두를 표시
plt.grid(which="major",axis='x',color="green",alpha=0.5,dashes=(5,2,1,2))
plt.grid(which="major",axis='y',color="red",alpha=0.5,linestyle="-.",linewidth=1)
#plt.grid(False)
```

```python
names = ['a','b','c']
values = [50,30,100]

plt.subplot(121) # 1행 2열의 subplot 중 왼쪽(1번) 위치
plt.bar(names,values)

plt.subplot(122) # 1행 2열의 subplot 중 왼쪽(2번) 위치
plt.barh(names,values)
```
