```python
# RNN이 생긴(필요한) 이유

# 월
# 화
# 수
# 목
# 금
# ==============

# 오늘

# ↓

# 비가

# ↓

# 온다

import numpy as np

# 데이터
x = np.array([1, 2, 3])
y = np.array([2, 3, 4])

# 초기 가중치
# Wx = np.random.randn()
# Wh = np.random.randn()
# b = np.random.randn()

Wx = 0.5   # 입력 가중치
Wh = 0.8   # 이전 상태 가중치
b  = 0.0   # 편향


# 초기 hidden state
h = 0

# 순전파 (forward)
outputs = []

for t in range(len(x)):
    h = np.tanh(Wx * x[t] + Wh * h + b)
    print(f"t={t+1}, h={h:.3f}")
    outputs.append(h)

# h1 = tanh(0.5*1 + 0.8*0)
#     = tanh(0.5)
#     ≈ 0.462

# h2 = tanh(0.5*2 + 0.8*0.462)
#     = tanh(1 + 0.3696)
#     = tanh(1.3696)
#     ≈ 0.878

# h3 = tanh(0.5*3 + 0.8*0.878)
#     = tanh(1.5 + 0.7024)
#     = tanh(2.2024)
#     ≈ 0.976

# 결과
# h1 ≈ 0.462
# h2 ≈ 0.878
# h3 ≈ 0.976

# 👉 단순히 1,2,3 넣은 게 아니라

# h2 = (현재 입력 2) + (이전 기억 h1)
# h3 = (현재 입력 3) + (이전 기억 h2)
```

```python
import numpy as np

# RNN 데이터
x = np.array(
    [
        [
            [1],[2],[3]
        ],
        [
            [2],[3],[4]
        ],
        [
            [3],[4],[5]
        ]
    ],dtype=np.float32
)

y = np.array(
    [
        [4],
        [5],
        [6]
    ],dtype=np.float32
)

print(x.shape)
print(y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Input

# 2️⃣ 핵심 개념 (딱 1줄)

# 👉 이전 상태(h)를 다음 계산에 계속 사용한다

# RNN 은 입력 3차원(data size, time_steps, features)
# (3, 3, 1)

model = Sequential([
    Input(shape=(3,1)),
    SimpleRNN(100),
    Dense(1)
])

model.summary()

model.compile(
    loss = "mse",
    optimizer = "adam"
)

model.fit(
    x,
    y,
    epochs = 1000,
    verbose=1
)
```

```python
preb = model.predict(x)
print(preb)

preb = model.predict(np.array([[[4],[5],[6]]]))
print(preb)
```
