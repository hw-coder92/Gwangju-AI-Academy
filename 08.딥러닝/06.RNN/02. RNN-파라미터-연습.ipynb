```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN , Dense , Input
```

```python
x = np.array([
    [1,2,3],
    [2,3,4],
    [3,4,5],
    [4,5,6],
])

y = np.array([4,5,6,7])

model = Sequential([
    Input(shape=(3,1)),
    SimpleRNN(2),
    Dense(1)
])

model.summary()

model.compile(
    loss = "mse",
    optimizer = "adam"
)

model.fit(
    x,
    y,
    epochs = 1000,
    verbose=1
)
```

```python
preb = model.predict(x)
print(preb)
```
