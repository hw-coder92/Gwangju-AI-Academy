```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
# x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])
```

```python
model = Sequential([
    Input(shape=(1,)),  # мһ…л Ҙ нҠ№м„ұ(feature)мқҙ 1к°ң
    Dense(1,activation="linear")
])

model.compile(optimizer="sgd",loss="mse")

model.fit(x, y, epochs=1000)
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x),"r")
plt.show()
```

```python
weights, bias = model.layers[0].get_weights()

print(weights)
print(bias)
```

```python
model.summary()
```

```python
# мҳҲмёЎ
pred = model.predict(np.array([10,11]))
print(pred)
```

```python
# - кІҪмӮ¬н•ҳк°•лІ• м§Ғм ‘ кө¬нҳ„ вҳ…x5

# мҳӨм°Ё кі„мӮ°
# вҶ“

# кё°мҡёкё° кі„мӮ°
# вҶ“

# к°ҖмӨ‘м№ҳ мҲҳм •
# вҶ“

# л°ҳліө
w = 10
lr = 0.1

for epoch in range(20):

    loss = (w-3)**2    
    grad = 2 * (w-3) # мң„мқҳ лҜёл¶„
    
    #мғҲ к°ҖмӨ‘м№ҳ = кё°мЎҙ к°ҖмӨ‘м№ҳ - н•ҷмҠөлҘ Г—кё°мҡёкё°
    w = w - lr * grad 

    print(epoch,round(w,3),round(loss,3))
```

```python
# лҚ°мқҙн„°

# л¬ём ң  мһҗлҸҷм°Ё мҶҚлҸ„мҷҖ м—°л№„

x = np.array([
    40,
    60,
    80,
    100,
    120
])

y = np.array([
    25,
    22,
    19,
    16,
    13
])

# н•ҙм•ј н•  мқј
# лӘЁлҚё мғқм„ұ
# н•ҷмҠө мҲҳн–ү
# мҶҚлҸ„ 90kmмқј л•Ң м—°л№„ мҳҲмёЎ
```

```python
model = Sequential([
    Input(shape=(1,)),  # м»¬лҹј(feature) к°ҜмҲҳ
    Dense(1,activation="linear")
])

model.compile(optimizer="adam",loss="mse")

model.fit(x, y, epochs=2000,verbose=1)
```

```python
# к°ҖмӨ‘м№ҳ л°”мқҙм–ҙмҠӨ нҷ•мқё
weights, bias = model.layers[0].get_weights()

print(weights,bias)

pred = model.predict(np.array([90]))
print(pred)
```
