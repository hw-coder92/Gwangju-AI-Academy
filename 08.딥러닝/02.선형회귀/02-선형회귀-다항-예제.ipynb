```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81,93,91,97])
```

```python
# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델을 설정합니다. 
# model.add(Dense(1, input_dim=2, activation='linear'))      --- 구형 버전
# model.add(Dense(1, input_shape=(2,), activation='linear'))
# model.add(Dense(1, Input(shape=(2,)), activation='linear')) 

model = Sequential([
    Input(shape=(2,)),  # 입력 특성(feature)이 2개
    Dense(1,activation="linear")
])

model.compile(optimizer="sgd",loss="mse")

model.fit(x, y, epochs=1000)
```

```python
weights, bias = model.layers[0].get_weights()

print(weights)
print(bias)
```

```python
model.summary()
```

```python
# 예측
pred = model.predict(np.array([[10, 5],[11,6]]))
print(pred)
```

```python
# 다항 문제
# 문제  광고비 + 영업사원수 → 매출

import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

# ==========================
# 데이터
# 광고비, 영업사원수
# ==========================

x = np.array([
    [10, 2],
    [20, 3],
    [30, 4],
    [40, 5]
])

# 매출
y = np.array([
    100,
    180,
    260,
    340
])

# 예측
# 광고비=35
# 영업사원수=4
```

```python
model = Sequential([
    Input(shape=(2,)),  # 컬럼 갯수
    Dense(1)
])

model.compile(optimizer="sgd",loss="mse")

model.fit(x, y, epochs=3000,verbose=0)
```

```python
# 가중치 확인
weights, bias = model.layers[0].get_weights()

print(weights)
print(bias)
```

```python
model.summary()
```

```python
# 예측
pred = model.predict(np.array([[35,4]]))
print(pred)
```
