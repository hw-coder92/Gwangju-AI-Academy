```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 데이터
# 1시간 => 50, 2시간 => 60

x = np.array([
    [1],
    [2],
    [3],
    [4],
    [5]
])

y = np.array([50,60,70,85,95])

# 학습
model = LinearRegression()
model.fit(x,y)

print("절편", model.intercept_)
print("가중치", model.coef_)
# 예측
pred = model.predict([[6],[7]])

print(pred)

plt.scatter(x,y)
plt.plot(x,model.predict(x))
```

```python
# y = x₁w₁ + x₂w₂ + x₃w₃ + b

# | 요소     | 의미          |
# | ------ | ----------- |
# | input  | 들어오는 정보     |
# | weight | 얼마나 중요하게 볼지 |
# | bias   | 기준값 조정      |
# | output | 최종 판단       |

inputs = [1,2,3]
weights = [0.3, 0.5, 0.2]

bias = 1
output = 0

for i in range(len(inputs)):
    output = output + (inputs[i]*weights[i])

output = output + bias
print(output)
```

```python
# y = x₁w₁ + x₂w₂ + x₃w₃ + b
# 넘파이 버전
inputs = np.array([1,2,3])
weights = np.array([0.3, 0.5, 0.2])

bias = 1

output = inputs @ weights 
print(output)
```

```python
# 입력 :
# 평수 = 30, 20, 50
# 역거리 = 10, 5, 10
# 연식 = 5, 4, 3

# 각각의 집 값을 계산하시오. => 넘파이 활용

# y = x₁w₁ + x₂w₂ + x₃w₃ + b

x = np.array([
    [30, 20, 50],
    [10, 5, 10],
    [5, 4, 3]
])

w = np.array([0.8, -0.3, -0.1])

print(w.shape)
output = x @ w

print(output)
```

```python
# 학생성적 예측

# 공부시간 10 , 5 , 8
# 출석률 95 , 90 , 98
# 결석횟수 2 , 5 , 1

#가중치

# 공부시간 → 3
# 출석률 → 0.5
# 결석횟수 → -2

# Numpy를 이용하여 각 학생의 성적 점수 계산하시오.

x = np.array([
    [10, 95, 2],  # 학생 1
    [5, 90, 5],   # 학생 2
    [8, 98, 1]    # 학생 3
])

w = np.array([3, 0.5, -2])

print(w.shape)
output = x @ w

print(output)
```

```python
# 활성화 함수의 이해

# Step
# =
# 꺼짐 / 켜짐

# Sigmoid
# =
# 0~1 사이 확률처럼 출력

# 활성화 함수
# =
# 뉴런에 비선형성을 추가해서
# 복잡한 문제를 풀게 만드는 장치

def step(x):

    if x >= 0:
        return 1
    
    return 0

for i in [-5,-1,0,1,5]:
    print(i,step(i))

def sigmoid(x):
    return 1/ (1 + np.exp(-x))

x = np.arange(-10,10,0.1)
y = sigmoid(x)

plt.plot(x,y)
```

```python
def perceptron(x1,x2):
    
    w1=1
    w2=1
    b=-1.5
    
    y = x1*w1 + x2*w2 + b

    if y > 0 :
         return 1

    return 0


data=[
(0,0),
(0,1),
(1,0),
(1,1)
]

for x1,x2 in data:
    print(x1,x2,perceptron( x1, x2 ) )
```

```python
from sklearn.linear_model import Perceptron

X=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

y = [0,1,1,0]

model = Perceptron()

# 내부동작

# 초기 상태

# w=랜덤
# b=랜덤
# w=[0.1 0.2]
# b=0

# 학습

# 입력
# ↓
# 예측
# ↓
# 정답 비교
# ↓
# 오차 계산
# ↓
# 가중치 수정
# 계속 반복

# 결국 이런 식의 뉴런을 찾음
# y = x₁w₁+x₂w₂+b
# 예)
# w=[1,1]
# b=-1.5

model.fit(
    X,
    y
)

print("가중치:")
print(model.coef_)

print()

print("편향:")
print(model.intercept_)

pred = model.predict(X)

print(pred)

# 예측 실패
# ↓

# XOR 해결 불가
# ↓

# 은닉층 필요
```
